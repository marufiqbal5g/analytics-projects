#1. Test the normality (Graphically and Numerically) of the variables Income and comment.

#Answer No 01

```r
# Graphical Normality Test: Histogram and Q-Q Plot
# Histogram
ggplot(data, aes(x = Income)) +
  geom_histogram(aes(y = ..density..), bins = 5, fill = "blue", color = "black") +
  geom_density(color = "red", linewidth = 1) +
  labs(title = "Histogram of Income", x = "Income", y = "Density")

# Q-Q Plot
ggplot(data, aes(sample = Income)) +
  stat_qq() +
  stat_qq_line(color = "red") +
  labs(title = "Q-Q Plot of Income", x = "Quantities", y = "Income")

# Numerical Normality Test: Shapiro-Wilk Test
shapiro_test <- shapiro.test(data$Income)
print(shapiro_test)
```

```r
# Load necessary library
library(ggplot2)

# Plot histogram
ggplot(data, aes(x=Income)) +
  geom_histogram(aes(y=..density..), bins=10, fill="blue", color="black", alpha=0.7) +
  geom_density(color="red") +
  labs(title="Histogram of Income", x="Income", y="Density")

# Plot Q-Q plot
qqnorm(data$Income, main="Q-Q Plot of Income")
qqline(data$Income, col = "red")
title(sub="x label = normal quantities, y label = dataset income")

# Perform Shapiro-Wilk test
shapiro.test(data$Income)
```

#2. Is the average income of customers in the dataset significantly different from $75 thousand?

#Answer No 02

```r
# Perform one-sample t-test
t_test_result <- t.test(data$Income, mu = 75)  # mu = 75 is the hypothesized mean

# Print the test results
print(t_test_result)
```

#3. Is the proportion of customers who accepted the personal loan in the last campaign greater than 40%?

#Answer No 03

```r
# Calculate the number of customers who accepted the personal loan
accepted_loans <- sum(data$Personal.Loan)

# Total number of customers
total_customers <- nrow(data)

# Perform one-sample proportion test
prop_test_result <- prop.test(accepted_loans, total_customers, p = 0.4, alternative = "greater")

# Print the test results
print(prop_test_result)
```

#4. Is there a significant difference in the average spending on credit cards (CCAvg) between customers who accepted the personal loan and those who did not?

#Answer No 04

```r
# Perform two-sample t-test
t_test_result <- t.test(CCAvg ~ Personal.Loan, data = data)

# Print the test results
print(t_test_result)
```

#5. Is there a significant difference in the average annual income between customers who accepted the personal loan and those who did not?

#Answer No 05

```r
# Perform two-sample t-test
t_test_result <- t.test(Income ~ Personal.Loan, data = data)

# Print the test results
print(t_test_result)
```

#6. Is there a significant difference in the proportion of customers who have a credit card versus those who do not.

#Answer No 06

```r
# Calculate the number of customers who have a credit card
have_credit_card <- sum(data$CreditCard)

# Calculate the number of customers who do not have a credit card
no_credit_card <- nrow(data) - have_credit_card

# Perform two-sample proportion test
prop_test_result <- prop.test(
  x = c(have_credit_card, no_credit_card),
  n = c(nrow(data), nrow(data)),
  alternative = "two.sided"
)

# Print the test results
print(prop_test_result)
```
